```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#ax = plt.axes(projection='3d')
from scipy.interpolate import griddata
```

```python
file_1000 = np.loadtxt('Coords_1000');
file_5000 = np.loadtxt('Coords_5000');
file_7500 = np.loadtxt('Coords_7500');
file_10000 = np.loadtxt('Coords_10000');
```

```python
def PLOT(file,name):
    x = file[:,0]
    y = file[:,1]
    z = file[:,2]
    plt.figure(figsize=(10,10))
    print("Global Min = ",np.min(z))
    plt.scatter(x,y,color = 'k')
    xlin = np.linspace(-20,20,20)
    ylin = np.linspace(-20,20,20)
    X, Y = np.meshgrid(xlin, ylin)
    Z = 1 + np.power(np.sin(X),2) + np.power(np.sin(Y),2) - 0.1 * np.exp(-X**2 - Y**2)
    plt.contour(X,Y,Z)
    plt.title("For Iteration = "+name)
    plt.savefig(name)
```

```python
PLOT(file_1000,"1000")
```

```python
PLOT(file_5000,"5000")
```

```python
PLOT(file_7500,"7500")
```

```python
PLOT(file_10000,"10000")
```

```python
def loadtxt(file):
    data   =  open(file, 'r') 
    result = []
    for index,value in enumerate(data):
        element  = value.split('.')
        if(len(element) == 1):
            continue
        else:
            temp = element[0]+'.' + element[1]   
        result.append(float(temp))
    return np.array(result)[:10000]
```

```python
MaxMinusMinForProccessor_1 = loadtxt("diff_1")
MinForProccessor_1         = loadtxt("min_1")
MaxMinusMinForProccessor_2 = loadtxt("diff_2")
MinForProccessor_2         = loadtxt("min_2")
MaxMinusMinForProccessor_4 = loadtxt("diff_4")
MinForProccessor_4         = loadtxt("min_4")
MaxMinusMinForProccessor_8 = loadtxt("diff_8")
MinForProccessor_8         = loadtxt("min_8")
```

```python
plt.plot(MaxMinusMinForProccessor_8,label = "Np=8")
plt.plot(MaxMinusMinForProccessor_4,label = "Np=4")
plt.plot(MaxMinusMinForProccessor_2,label = "Np=2")
plt.plot(MaxMinusMinForProccessor_1,label = "Np=1")
plt.xlabel('Iteration')
plt.ylabel('Max-Min')
plt.legend()
#plt.savefig("Max-min.png")
```

```python
plt.plot(MinForProccessor_8,label = "Np=8")
plt.plot(MinForProccessor_4,label = "Np=4")
plt.plot(MinForProccessor_2,label = "Np=2")
plt.plot(MinForProccessor_1,label = "Np=1")
plt.xlabel('Iteration')
plt.ylabel('Min')
plt.legend()
#plt.savefig("MIN.png")
```
