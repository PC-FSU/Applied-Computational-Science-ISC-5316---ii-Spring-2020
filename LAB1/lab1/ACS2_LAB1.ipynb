```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
from scipy import stats
import scipy
```

# Part 1,Section 1

```python
Random_Number = np.random.rand(100000)
```

 # Part 1,Section 2

```python
_ = plt.hist(Random_Number,bins = 50)
plt.xlabel('Numbers Generated')
plt.ylabel('Frequency of Number')
plt.savefig('Random Number')
```

To check the property of random data, like mean and standard deviation.

```python
New_list = pd.Series(Random_Number)
```

```python
New_list.describe()
```

# Part 1, Section 3 and 4

```python
inverted_data = norm.ppf(New_list, loc=0, scale=np.sqrt(2))
_=plt.hist(inverted_data,bins = 50)
plt.xlabel('Numbers Generated')
plt.ylabel('Frequency of Number')
plt.savefig('Number Generated using inverse sampling')
```

To check the property of data obtained by inverse sampling, like mean and standard deviation.

```python
pd.Series(inverted_data).describe()
```

# Part 1, Section 5

```python
_,_ = stats.probplot(inverted_data,plot=plt)
plt.savefig('QQ-Plot')
```

 # Part 1, Section 6

```python
scipy.stats.kstest(inverted_data,'norm',args = (0,np.sqrt(2)))
```

# Part 2, Section 3

```python
Number_file = ['Number_1000.txt','Number_10000.txt','Number_100000.txt']
Frequency_files = ['Observe_frequency_1000.txt','Observe_frequency_10000.txt','Observe_frequency_100000.txt']
for number_file,frequency_file in zip(Number_file,Frequency_files):
    number = np.loadtxt(number_file)
    number = np.reshape(number,(1,-1))
    Observe_frequency = np.loadtxt(frequency_file)
    Expected_frequency = [Observe_frequency[-1] for i in range(len(Observe_frequency)-1)]
    Observe_frequency = Observe_frequency[:-1]
    print(stats.chisquare(Observe_frequency,Expected_frequency))
```
