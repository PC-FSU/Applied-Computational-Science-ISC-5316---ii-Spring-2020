### Problem 1, Part 3
<br> For this problem,we will find area under tail for different, which give the proportion of rejection.<br>
We reject null hypothesis if probability doesn't lie within $2\sigma$. <br>
$2\sigma = {\sqrt \frac{p(1-p)}{n} }$ = $\frac{1}{\sqrt n}$ if p = 0.5.<br>
Therefore probability of rejecting is
$$P(p^{*} < 0.5 - \frac{1}{\sqrt n} | p = 0.55)  + P(p^{*} > 0.5 + \frac{1}{\sqrt n} | p = 0.55)$$<br>
$$P(Z<\frac{(0.5 - \frac{1}{\sqrt n}) - 0.55}{\frac{0.5}{\sqrt n}}) + P(Z > \frac{(0.5 + \frac{1}{\sqrt n}) - 0.55}{\frac{0.5}{\sqrt n}}) $$<br>
$$P(Z < \sqrt n - 2 - 1.1\sqrt n ) + P(Z > \sqrt n + 2 - 1.1\sqrt n )$$
<br>
Where Z is the score. 

```python
# The Z-score can calculated from Z-score tabel.
n = [50,100,250,500,1000]
a =  [0.035,0.013,0,0,0]
b = [0.0985,0.1587,0.34,1-0.4129,1-0.1230]
```

```python
from operator import add
theoretical_power = list(map(add, a, b))
```

### Problem 2,Part 2

```python
def likelihood(k,lamb):
    X = np.linspace(2,5,8,endpoint = True)
    temp = 1
   # X=[3]
    for x in X:
        temp = temp*(k/lamb)*np.power((x/lamb),k-1)*np.exp(-np.power((x/lamb),k))
    return temp
```

### Problem 2, Part 3 and Part 4

```python
def log_likelihood(X,param):
    '''
    Input argument : 
    X : Array of point X
    param : param[0] = k, param[1] = lambda
    Output:
    Log_liklihood for input array X, for a given k and lambda
    '''
    k = param[0]
    lamb = param[1]
    temp = 0
    for x in X:
         temp += np.log((k/lamb)*np.power((x/lamb),k-1)*np.exp(-np.power((x/lamb),k)))
    return temp
```

```python
def log_likelihood_derivative(X,param):
    '''
    Input argument : 
    X : Array of point X
    param : param[0] = k, param[1] = lambda
    Output:
    Derivative of log_likelihood function w.r.t to k and lambda
    '''
    k = param[0]
    lamb = param[1]
    # for derivative w.r.t K
    temp = 0
    # for derivative w.r.t lambda
    temp1 = 0
    for x in X:
        temp +=  (-np.power((x/lamb),k)*np.log(x/lamb) + np.log(x/lamb) + 1/k)
        temp1 +=  (k/lamb)*( np.power(x/lamb,k) - 1 )
    res = np.array((temp,temp1))
    return res
```

```python
def plot_liklihood(k,lamb):
    '''
    Input argument : 
    k and lambda
    Output:
    plot of weibull distribution
    '''
    X = np.linspace(0,8,50)
    temp = k/lamb
    temp1 = X/lamb
    Y_axis = temp*np.power(temp1,k-1)*np.exp(-np.power(temp1,k))
    X_axis = X
    plt.plot(X_axis,Y_axis)
    plt.xlabel('X')
    plt.ylabel =('Likelihood')
    plt.show()
```

**for X in range (3,6)**

```python
param = [3,3]
X = np.linspace(3,6,8,endpoint = True)
#results = root(derivative, param,args=X,method ='hybr',options={'disp': True,'eps': 1e-5})
results = minimize(log_likelihood, param,args=X,method ='BFGS',jac = log_likelihood_derivative,options={'disp': True,'eps': 1e-8})
k,lamb=results.x[0],results.x[1]
print('k,lamb = ',k,lamb)
plot_liklihood(k,lamb)
```
