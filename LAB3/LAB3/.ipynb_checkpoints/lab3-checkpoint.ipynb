```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.optimize
```

```python
def rho (x, k, y) :
    return (k / y) * (x / y) ** (k - 1) * np.exp(-(x / y) ** k)

def L (x, ki, yi):
    prod = 1
    for xi in x:
        prod *= rho(xi, ki, yi)
    return prod

def logL (x, ki, yi):
    return sum([np.log(rho(xi, ki, yi)) for xi in x])
```

```python
# contour plot of likelihood function
y = np.linspace(2, 8)
k = np.linspace(0.1, 5)
x = [2, 5]
plt.contour(k, y, [[L(x, ki, yi) for ki in k] for yi in y], 10)
plt.ylabel(r"$\lambda$")
plt.xlabel(r"$k$")
```

```python
# test optimization -- looking for approx k=2.5, lambda=4 by eyeballing contour plot above
scipy.optimize.minimize(lambda params : -logL(x, params[0], params[1]), [3, 3])
```
