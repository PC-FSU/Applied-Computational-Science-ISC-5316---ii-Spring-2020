```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Text_list = ['newton.txt','Stepest.txt','BFGS.txt']
label_list = ['newton','Stepest','BFGS']
```

```python
for index,file in enumerate(Text_list):
    temp=0
    temp=np.loadtxt(file)
    x2=0
    x1=0
    x2 = temp[::2]
    x1 = temp[1::2]
    f = 100*np.power((x2-np.power(x1,2)),2) +  np.power(1-x1,2)
    x_axis = np.linspace(0,len(x1),len(x1))
    plt.loglog(x_axis,f,label = label_list[index])
    plt.xlabel('Iteration')
    plt.ylabel('Function Value')
    plt.title('Convergence')
    plt.legend()
plt.savefig('Convergence')
```
