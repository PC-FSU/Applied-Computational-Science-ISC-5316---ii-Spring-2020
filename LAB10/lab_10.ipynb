```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_4 = np.loadtxt("4.txt")
file_8 = np.loadtxt("8.txt")
file_16 = np.loadtxt("16.txt")
```

```python
plt.loglog(file_4[1:],label = "n=4")
plt.loglog(file_8[1:],label = "n=8")
plt.loglog(file_16[1:],label = "n=16")
plt.legend()
plt.ylabel('Residual')
plt.xlabel('Iterations')
plt.title('Log-Log plot FDE')
plt.savefig('1')
```

```python
file_4b = np.loadtxt("4b.txt")
file_8b = np.loadtxt("8b.txt")
file_16b = np.loadtxt("16b.txt")
```

```python
plt.loglog(file_4b[1:],label = "n=4")
plt.loglog(file_8b[1:],label = "n=8")
plt.loglog(file_16b[1:],label = "n=16")
plt.legend()
plt.ylabel('Residual')
plt.xlabel('Iterations')
plt.title('Log-Log plot FDE with periodic boundary')
plt.savefig('2')
```

```python
file_4b_res  = np.loadtxt("4bres.txt").T
file_8b_res  = np.loadtxt("8bres.txt").T
file_16b_res = np.loadtxt("16bres.txt").T
file_list = [file_4b_res,file_8b_res,file_16b_res]
N = [4,8,16]
```

```python
for index,value in enumerate(N):
    xlist = np.linspace(0, 1, value+1)
    ylist = np.linspace(0, 1, value+1)
    X, Y = np.meshgrid(xlist, ylist)
    Z = file_list[index]
    fig,ax=plt.subplots(1,1)
    cp = ax.contourf(X, Y, Z)
    fig.colorbar(cp) # Add a colorbar to a plot
    ax.set_title('Filled Contours Plot')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    plt.show()
```
